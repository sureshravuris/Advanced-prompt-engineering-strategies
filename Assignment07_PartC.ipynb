#### Defining Various Functions

**List of Fields**

```python
field_list = ['Human Resources', 'Teaching', 'Finance', 'Healthcare', 'Marketing', 'Legal', 'Engineering', 'Information Technology', 'Art and Design', 'Customer Service']
```

**Function to Test Prompts**

```python
# Function to test prompts
def test_prompt(prompt, model="gpt-3.5-turbo", max_tokens=500):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,
        temperature=0  # Set temperature to 0 for deterministic output
    )
    return response.choices[0].message.content.strip()
```

**Get Prompts and Outputs for a specific Field**

```python
# Function to get prompts and outputs for a specific field
def get_prompts_and_outputs(field):
    if field in prompts:
        outputs = []
        for item in prompts[field]:
            prompt_text = item['prompt']
            output_text = test_prompt(prompt_text)
            outputs.append({'prompt': prompt_text, 'output': output_text})
        return outputs
    else:
        print(f"Field '{field}' not found.")
        return None
```

**Pretty-Print Prompts and Outputs**

```python
# Function to pretty-print the prompts and outputs
def pretty_print_prompts_and_outputs(field_number):
    field = field_list[field_number]
    outputs = get_prompts_and_outputs(field)
    if outputs:
        print(f"Prompts and Outputs for Field {field}\n\n")
        for idx, item in enumerate(outputs):
            print(f"Prompt {idx + 1}:\n")
            print(item['prompt'])
            print("\nOutput:\n")
            print(item['output'])
            print("\n" + "-" * 50 + "\n")
    else:
        print("No prompts and outputs to display.")
```
